```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('seaborn')

empl_sector_per_file = 'datasets/share-of-employment-in-agriculture-industry-and-services.csv'
empl_sector_pop_file = 'datasets/employment-by-economic-sector.csv'
sector_gdp_file = 'datasets/composition-of-national-gross-domestic-product-by-sector.csv'

empl_per = pd.read_csv(empl_sector_per_file)
empl_pop = pd.read_csv(empl_sector_pop_file)
sect_gdp = pd.read_csv(sector_gdp_file)
```

```python
empl_per.head()
```

```python
empl_per.size
```

```python
empl_per_usa = empl_per.loc[(empl_per['Entity'] == 'United States of America') | (empl_per['Entity'] == 'United States')].copy()
empl_per_usa
```

```python
empl_per.columns
```

```python
empl_per_usa['Share of workers in financial sector (GGDC, 2017)'] = empl_per_usa['Share of workers in financial sector (GGDC, 2017)'] * 100
empl_per_usa
```

```python
empl = ['Employment in services (% of total employment) (modeled ILO estimate)', 'Employment in industry (% of total employment) (modeled ILO estimate)', 'Employment in agriculture (% of total employment) (modeled ILO estimate)', 'Share of workers in financial sector (GGDC, 2017)']
empl_per_usa.plot(kind='line', x='Year', y=empl, xlim=1990, figsize=(18,8))

plt.title('Employment trends in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')
plt.show()
```

```python
empl_per_usa.plot(kind='line', x='Year', y=['Employment in services (% of total employment) (modeled ILO estimate)'],figsize=(18,8))

plt.title('Service sector employment in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')

plt.axvline(x=2008, color='black', linestyle='dotted')
plt.text(x=2004, y=79, s="2008 Financial Market Crash")

plt.show()
```

```python
empl_per_usa.plot(kind='line', x='Year', y=['Employment in industry (% of total employment) (modeled ILO estimate)'], figsize=(18,8))

plt.title('Industrial sector employment in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')

plt.axvline(x=2008, color='black', linestyle='dotted')
plt.text(x=2008, y=21, s="    2008 Financial Market Crash")

plt.show()
```

```python
empl_per_usa.plot(kind='line', x='Year', y=['Employment in agriculture (% of total employment) (modeled ILO estimate)'],figsize=(18,8))

plt.title('Agricultural sector employment in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')

#plt.axvline(x=2008, color='black', linestyle='dotted')
#plt.text(x=2004, y=79, s="2008 Financial Market Crash")

plt.show()
```

```python
empl_per_usa.plot(kind='line', x='Year', y=['Share of workers in financial sector (GGDC, 2017)'], figsize=(18,8))

plt.title('Financial sector employment in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')

#plt.axvline(x=2008, color='black', linestyle='dotted')
#plt.text(x=2000, y=8, s="2008 Financial Market Crash")

plt.show()
```
