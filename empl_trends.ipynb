```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('seaborn')
```

```python
empl_file_path = 'datasets/share-of-employment-in-agriculture-industry-and-services.csv'

df = pd.read_csv(empl_file_path)
df.head()
```

```python
df.size
```

```python
df_usa = df.loc[(df['Entity'] == 'United States of America') | (df['Entity'] == 'United States')]
df_usa
```

```python
df.columns
```

```python
df_usa['Share of workers in financial sector (GGDC, 2017)'] = df_usa['Share of workers in financial sector (GGDC, 2017)'] * 100
df_usa
```

```python
empl = ['Employment in agriculture (% of total employment) (modeled ILO estimate)', 'Employment in industry (% of total employment) (modeled ILO estimate)', 'Employment in services (% of total employment) (modeled ILO estimate)']
df_usa.plot(kind='line', x='Year', y=empl, figsize=(18,8))

plt.title('Employment trends in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')
plt.show()
```

```python
df_usa.plot(kind='line', x='Year', y=['Share of workers in financial sector (GGDC, 2017)'], figsize=(18,8))

plt.title('Financial sector employment trends in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')

plt.axvline(x=2008, color='black', linestyle='dotted')
plt.text(x=2000, y=8, s="2008 Financial Market Crash")

plt.show()
```

```python
df_usa.plot(kind='line', x='Year', y=['Employment in services (% of total employment) (modeled ILO estimate)'], figsize=(18,8))

plt.title('Service sector employment trends in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')

plt.axvline(x=2008, color='black', linestyle='dotted')

plt.text(x=2004, y=79, s="2008 Financial Market Crash")

plt.show()
```

```python
df_usa.plot(kind='line', x='Year', y=['Employment in industry (% of total employment) (modeled ILO estimate)'], figsize=(18,8))

plt.title('Industry sector employment trends in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')

plt.axvline(x=2008, color='black', linestyle='dotted')

plt.text(x=2008, y=21, s="    2008 Financial Market Crash")

plt.show()
```
