# Analyzing Global Trends of Employment Sectors
### Dataset provided by [Our World In Data](https://ourworldindata.org/grapher/share-of-employment-in-agriculture-industry-and-services)

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('seaborn')
```

```python
empl_file_path = 'datasets/share-of-employment-in-agriculture-industry-and-services.csv'

df = pd.read_csv(empl_file_path)
df.head()
```

```python
df.size
```

```python
df_usa = df.loc[(df['Entity'] == 'United States of America') | (df['Entity'] == 'United States')]
df_usa
```

```python
df.columns
```

```python
empl = ['Employment in agriculture (% of total employment) (modeled ILO estimate)', 'Employment in industry (% of total employment) (modeled ILO estimate)', 'Employment in services (% of total employment) (modeled ILO estimate)']
df_usa.plot(kind='line', x='Year', y=empl, figsize=(18,8))

plt.title('Employment trends in the USA')
plt.xlabel('Year')
plt.ylabel('% of total employment')
plt.show()
```
